```python
from sklearn.linear_model import LinearRegression
```

```python
reg1=LinearRegression()
reg1.fit(df2[['x']],df2['y'])
```

```python
reg1.score(df2[['x']],df2[['y']])
```

```python
plt.figure(figsize=(12,5))
plt.scatter(df2[['x']],df2[['y']],color='#004B95')
plt.plot(df2[['x']],reg1.predict(df2[['x']]) ,color='yellow' , label='Best Fit Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Non Linear Relationship')
plt.legend()
```

```python
reg2=LinearRegression()
reg2.fit(df[['x']],df.y)
```

```python
reg2.score(df[['x']],df.y)
```

```python
reg2.score(df[['x']],df.y)*100
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(9, 6))
plt.scatter(df['x'], df['y'], color='Red')
plt.plot(df['x'], reg2.predict(df[['x']]), color='Green', label='Best Fit Line')
plt.xlabel('x')
plt.ylabel('y')

plt.title('Non Linear Relationship')
plt.legend()

plt.tight_layout()
plt.show()
```
