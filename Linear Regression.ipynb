```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(df[['area']],df[['rent']])
```

```python
coefficient_m=reg.coef_
```

```python
reg.coef_
```

```python
# Define the variable coefficient
coefficient_m = reg.coef_[0]
```

```python
coefficient_m
```

```python
intercept=reg.intercept_
intercept
```

```python
reg.predict([[2000]])
```

```python
coefficient_m*2000+intercept
```

```python
pred=reg.predict(df[['area']])
pred
```

```python
df['pred']=reg.predict(df[['area']])
df.head()
```

```python
plt.plot(df['area'], reg.predict(df[['area']]))
plt.xlabel('Area')
plt.ylabel('Rent')
plt.title('Dhaka City Rent')
plt.scatter(df['area'], df['rent'], color='red', marker='o')
```
