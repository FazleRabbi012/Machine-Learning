```python
inputs=df.drop('Seat_Type',axis='columns')
inputs
```

```python
target=df['Seat_Type']
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python

le_Institute=LabelEncoder()
le_Quota=LabelEncoder()
le_Gender=LabelEncoder()
le_Year=LabelEncoder()
le_Academic_Program_Name=LabelEncoder()
le_Closing_Rank=LabelEncoder()
le_Opening_Rank=LabelEncoder()

```

```python
inputs['Institute_n']=le_Institute.fit_transform(inputs['Institute'])
inputs['Quota_n']=le_Quota.fit_transform(inputs['Quota'])
inputs['Gender_n']=le_Gender.fit_transform(inputs['Gender'])
inputs['Year_n']=le_Year.fit_transform(inputs['Year'])
inputs['Academic_Program_Name_n']=le_Academic_Program_Name.fit_transform(inputs['Academic_Program_Name'])
inputs['Opening_Rank_n']=le_Year.fit_transform(inputs['Opening_Rank'])
inputs['Closing_Rank_n']=le_Year.fit_transform(inputs['Closing_Rank'])
inputs
```

```python
inputs_n=inputs.drop(['Institute','Quota','Academic_Program_Name','Year','Gender','Closing_Rank','Opening_Rank'],axis='columns')
```

```python
inputs_n
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier(criterion='entropy')
```

```python
model.fit(inputs_n,target)
```

```python
print(inputs_n.dtypes)
```

```python
model.score(inputs_n,target)
```

```python
model.predict([[4,0,1,0,149,1646,391]])
```
