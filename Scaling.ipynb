```python
li= [250, 300, 400, 350, 200, 500, 550, 650, 650, 330, 470, 320, 600, 430, 950, 368, 234, 869, 240 ]
```

```python
li.sort()
li
```

```python
import pandas as pd
df=pd.DataFrame(li,columns=["cost"])
df
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mmn=MinMaxScaler()
```

```python
df['cost_Normalization']=mmn.fit_transform(df[['cost']])
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
df['cost_Standardization']=ss.fit_transform(df[['cost']])
df.head()
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft=FunctionTransformer()
df['Cost_logTransformation']=ft.fit_transform(df[["cost"]])
df.head()
```

```python
from sklearn.preprocessing import MaxAbsScaler
mas=MaxAbsScaler()
df['cos_MaxAbsScaler']=mas.fit_transform(df[["cost"]])
df.head()
```
